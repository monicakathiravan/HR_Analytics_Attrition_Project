# HR Analytics Project: Employee Attrition Analysis

This project aims to analyze employee attrition data to identify patterns and key factors that influence employee turnover. The dataset includes attributes such as employee demographics, job satisfaction, and work history. Through data cleaning, exploration, and analysis, the goal is to understand the causes of attrition and provide actionable insights to improve employee retention.

```python
import pandas as pd

df = pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')

df.head()
```

## Step 1: Data Cleaning

In this step, we'll clean the dataset by addressing the following:
- Checking for missing values.
- Removing any duplicate rows.
- Verifying that the data types are correct.
- Identifying and handling any zero or negative values.

```python
missing_values = df.isnull().sum()
print("Missing values in each column:")
print(missing_values)

duplicate_count = df.duplicated().sum()
print(f"\nNumber of duplicate rows: {duplicate_count}")

print("\nData types of each column:")
print(df.dtypes)

numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
for col in numeric_cols:
    if (df[col] <= 0).any():
        print(f"Column '{col}' contains zero or negative values.")
```

## Step 2: Exploratory Data Analysis (EDA)

In this step, we'll perform some basic exploratory data analysis to understand the distribution of values and the relationships between various features.

```python
print("\nSummary statistics for numerical columns:")
print(df.describe())

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,6))
sns.histplot(df['Age'], kde=True)
plt.title('Age Distribution')
plt.show()
```

## Step 3: Insights and Conclusion

After analyzing the data, we can draw several conclusions:
- **Age distribution**: Most employees in the dataset are between 25 and 45 years old.
- **Attrition Rate**: The attrition rate is higher among employees in certain departments and with fewer years at the company.
- **Job Satisfaction**: Employees with lower job satisfaction levels tend to leave the company more often.

These insights could help the organization improve its retention strategies.
